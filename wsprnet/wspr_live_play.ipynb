```python
class SpotLoader(ezr.pickle_cache_mixin):
    pkc = ezr.pickle_cache_state('active')
    
    def __init__(self):
        self.query = """
            SELECT
                *
            FROM
                wspr.rx
            WHERE
                time > '2021-01-01'
            AND
                match(rx_sign, 'N4UZZ.*')
            LIMIT 10
        """
        
    async def _get_frame_from_db(self):
        s = ClientSession()
        client = ChClient(s, url="https://db1.wspr.live")
        if not await client.is_alive():
            raise RuntimeError("database conneciton failed!")

        all_rows = await client.fetch(r"""
            SELECT
                *
            FROM
                wspr.rx
            WHERE
                time > '2021-01-01'
            AND
                match(rx_sign, 'N4UZZ.*')
        """)  
        df = pd.DataFrame(all_rows)   
        await client.close()
        await s.close()
        return df
    
    @ezr.pickle_cached_container()
    def df_raw(self):
        return asyncio.run(self._get_frame_from_db())
        
        
loader = SpotLoader()
loader.df_raw
    
```
