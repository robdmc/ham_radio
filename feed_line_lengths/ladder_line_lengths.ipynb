```python
from astropy import units as u
from astropy import constants as con
import numpy as np
import pandas as pd
```

```python
freqs = np.array([
    np.mean([3.5, 4.0]),
    np.mean([7, 7.3]),
    np.mean([14.1, 14.35]),
    np.mean([21, 21.45]),
    np.mean([28.0, 29.7])
    
]) * u.MHz

vf = .91
c = vf * con.c
freqs
```

```python
base_lengths = (c / freqs).to(u.imperial.ft) / 8.
df = pd.DataFrame({'base_length': base_lengths.value})
df.index = [80, 40, 20, 15, 10]
df['delta'] = df.base_length / 2
for nn in range(2, 31):
    col = f'x{nn:02d}'
    df[col] = nn * df.base_length
df = df.T

# df = df.drop(['base_length', 'delta'])

df_delta = (df - 62).abs()
for col in df_delta.columns:
    df_delta.loc[:, col] = df_delta.loc[df_delta[col] == df_delta[col].min(), col]
# df_delta = df[df_delta < 2]
df_delta
```
